```python
def map_cp_dist(r):
    s = len(r) - 1
    cps = []
    while s > 0:
        pr = r[s, :s+1]
        s = s - np.argmax(pr)
        cps.append(s)
    return list(reversed(cps))
```

```python
map_cp_dist(rs.T)
```

```python
def mc_cp_dist(r, n=1000):
    """Walk backwards through the sequence, via mc"""
    bins = np.zeros(len(r), dtype=np.float64)
    for _i in range(n):
        s = len(r) - 1
        while s > 0:
            pr = r[s, :s+1]
            s = s - np.random.choice(range(s+1), p=pr)
            bins[s] += 1
    return bins / n
```
